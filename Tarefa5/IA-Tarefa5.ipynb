# Árvore de Decisão

1.Importação de Dados

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import classification_report
```

```python

X, y = load_iris(return_X_y=True, as_frame=True)
```

```python
# Remove de X as colunas petal length (cm) e petal width (cm)

X = X.drop(['sepal length (cm)', 'sepal width (cm)'], axis=1)
```

```python
# Divisão de Dados em Treino e Teste

X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
# Instanciação do Modelo DecisionTreeClassifier()

model = tree.DecisionTreeClassifier()
```

```python
# Treino do Modelo com Dados de Treino

model.fit(X_treino, y_treino)
```

```python
# Visualizando os Dados

import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter(X_treino['petal length (cm)'],
           X_treino['petal width (cm)'],
           c=y_treino)

ax.set_xlabel('petal length (cm)')
ax.set_ylabel('petal width (cm)')

plt.show()
```

```python
yte_prediction = model.predict(X_teste)
```

```python
model.score(X_teste, y_teste)
```

```python
# Visualizando a Árvore

fig, ax = plt.subplots()

tree.plot_tree(model)

plt.show()
```
