```python
# Importação das bibliotecas
from PIL import Image
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
def load_dataset():
  # Passo 1: Definir os caminhos para 5 imagens
  image_paths = [
      "/content/drive/My Drive/datasets/imagem1.jpg",
      "/content/drive/My Drive/datasets/imagem2.jpg",
      "/content/drive/My Drive/datasets/imagem3.jpg",
      "/content/drive/My Drive/datasets/imagem4.jpg",
      "/content/drive/My Drive/datasets/imagem5.jpg"
  ]

  # Passo 2: Preparar arrays para features (X) e rótulos (y)
  X = []  # Para armazenar as imagens processadas
  y = [1, 0, 0, 0, 1]  # Rótulos fictícios associados às imagens

  for path in image_paths:
    # Carregar e redimensionar a imagem para 64x64
    img = Image.open(path).resize((64, 64))

    # Converter para array NumPy
    img_array = np.array(img)

    # Achatar a matriz de pixels para uma dimensão (12288,1)
    flat_array = img_array.flatten()

    # Normalizar os valores para o intervalo [0, 1]
    flat_array = flat_array / 255.0

    # Adicionar ao conjunto de treinamento
    X.append(flat_array)

  # Converter X para um array NumPy para compatibilidade com scikit-learn
  X = np.array(X)

  return X, y
```

```python
# Passo 3: Carregar o dataset

X, y = load_dataset()
```

```python
# Passo 4: Padronizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Passo 5: Treinar o modelo de regressão logística
model = LogisticRegression()
model.fit(X_scaled, y)
```

```python
# Passo 6: Testar no próprio conjunto de Treino
predictions = model.predict(X_scaled)
print("Predições:", predictions)
print("Rótulos reais:", y)
```
